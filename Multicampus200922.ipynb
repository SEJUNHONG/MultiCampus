```python
train.columns
fn=['Gender', 'Age_mean', 'Embarked_S', 'Embarked_C', 'Embarked_Q', 'Family_S', 'Family_M', 'Family_L'] # 독립변수들 (입력 데이터)
xTrain=train[fn]
xTrain.head()
yLabel=train['Survived'] # 종속변수 (출력데이터)
```

```python
# 문제 : xTrain, 정답 : yLabel -> 모델링 (의사결정트리 알고리즘) => 모델
# 테스트 입력 데이터 : xTest -------------------------------------> 입력 ====> 생존여부 출력 (예측결과)
```

```python
xTest=test[fn]
```

```python
# 모델링
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(max_depth=3, random_state=2020)
model
```

```python
model.fit(xTrain, yLabel)
```

```python
prediction=model.predict(xTest)
prediction.shape
```

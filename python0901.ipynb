```python
def simPearson(data, name1, name2):
    sumX=0
    sumY=0
    sumPowX=0
    sumPowY=0
    sumXY=0
    count=0
    for i in data[name1]:
        if i in data[name2]:
            sumX+=data[name1][i]
            sumY+=data[name2][i]
            sumPowX+=pow(data[name1][i], 2)
            sumPowY+=pow(data[name2][i], 2)
            sumXY+=data[name1][i]*data[name2][i]
            count+=1
    r=(sumXY-(sumX*sumY/count))/sqrt((sumPowX-pow(sumX, 2)/count)*(sumPowY-pow(sumY,2)/count))
    return r
```

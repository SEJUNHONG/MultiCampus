```python
df=pd.DataFrame(np.random.randn(4, 3), columns=list('bde'), index=['Seoul', 'Busan', 'Daegu', 'Incheon'])
df
```

```python
f=lambda x:x.max()-x.min()
df.apply(f)

df.apply(f, axis=0)
df.apply(f, axis='index')
```

```python
f2=lambda x:x.mean()
df.apply(f2, axis=1)
```

```python
df.mean(axis=0)
df.mean(axis=1)
```

```python
def f3(x):
    return pd.Series([x.min(), x.max()], index=['min','max'])
df.apply(f3)
```

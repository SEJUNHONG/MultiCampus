```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("train.csv")
```

```python
train
```

```python
train.describe()
```

```python
train.isnull().sum().sort_values()[-6:]
```

```python
train['LotFrontage'].value_counts()
```

```python
mean_LotFrontage=round(train[train['LotFrontage'].notnull()]['LotFrontage'].mean())
```

```python
train['LotFrontage']=np.where(train['LotFrontage'].isnull(), mean_LotFrontage, train['LotFrontage'])
```

```python
train['LotFrontage']
```

```python
train['FireplaceQu']
```

```python
train['FireplaceQu'].value_counts()
```

```python
train['FireplaceQu'].fillna(method='ffill')
```

```python
train['Fence']
```

```python
train['Fence'].value_counts()
```

```python
train['Fence'].fillna(method='bfill')
```

```python
train['Alley']
```

```python
train['Alley'].value_counts()
```

```python
train['Alley']=np.where(train['Alley'].isnull(), np.where(train['Alley'].index % 2 ==0, 'Grvl', 'Pave'), train['Alley'])
```

```python
train['Alley']
```

```python
train['MiscFeature']
```

```python
train['MiscFeature'].value_counts()
```

```python
train['MiscFeature'].fillna(method='bfill')
```

```python
train['PoolQC']
```

```python
train['PoolQC'].value_counts()
```

```python
train=train.drop(['PoolQC', 'Alley','MiscFeature'], axis=1)
```

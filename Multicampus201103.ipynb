```python
from keras.datasets import imdb
```

```python
(trainData, trainLabel), (testData, testLabel)=imdb.load_data(num_words=10000)
```

```python
max([max(seq) for seq in trainData])
```

```python
word_index=imdb.get_word_index()
```

```python
rev_word_index=dict([(value, key) for (key, value) in word_index.items()])
```

```python
decodedReview=' '.join([rev_word_index.get(i-3, '?') for i in trainData[0]])
decodedReview
```

```python
def vs(seq, dimension=10000):
    res=np.zeros((len(seq), dimension))    
    for i, sequence in enumerate(seq):
        res[i, sequence]=1
    return res

xTrain=vs(trainData)
xTest=vs(testData)
```

```python
yTrain=trainLabel.astype('float32')
yTest=testLabel.astype('float32')
```

```python
model=Sequential()
model.add(Dense(32,input_shape=(10000,) ,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
xVal=xTrain[:10000]
partialXTrain=xTrain[10000:]

yVal=yTrain[:10000]
partialYTrain=yTrain[10000:]
```

```python
history=model.fit(partialXTrain, partialYTrain, epochs=20,
         batch_size=512, validation_data=(xVal, yVal))
```

```python
historyDict=history.history
```

```python
historyDict.keys()
```

```python
loss=history.history['loss']
accuracy=history.history['accuracy']
val_loss=history.history['val_loss']
val_accuracy=history.history['val_accuracy']
```

```python
import matplotlib.pyplot as plt
```

```python
epochs=range(1, len(accuracy)+1)
epochs
```

```python
plt.plot(epochs, loss, 'bo', label='training cost')
plt.plot(epochs, val_loss, 'b', label='validation cost')
plt.legend()
plt.show()
```

```python
plt.plot(epochs, accuracy, 'bo', label='training cost')
plt.plot(epochs, val_accuracy, 'b', label='validatioin cost')
plt.legend()
plt.show()
```

```python
from urllib.request import urlopen
from urllib.parse import quote_plus
```

```python
baseUrl="https://search.naver.com/search.naver?where=image&sm=tab_jum&query="
word=input("검색어를 입력하세요 : ")
num=int(input("크롤링할 데이터 개수 : "))
url=baseUrl+quote_plus(word)
html=urlopen(url)
print(html)
```

```python
soup=BeautifulSoup(html, "html.parser")
print(soup)
```

```python
img=soup.find_all(class_="_img")
```

```python
n=1
for i in img:
    imgUrl=i['data-source']
    img=urlopen(imgUrl).read()
    f=open("img/down2"+str(n)+".jpg", "wb")
    f.write(img)
    f.close()
    n+=1
    if num < n:
        break
```

```python
n=1
for i in img:
    imgUrl=i['data-source']
    with urlopen(imgUrl) as f:
        with open("img/naver"+str(n)+".jpg", "wb") as na:
            img=f.read()
            na.write(img)
    n+=1
    if n>num:
        break
```

```python
def get_conten(driver):
    html=driver.page_source
    soup=BeautifulSoup(html, "lxml")
    try:
        content=soup.select("div.C4VMK > span")[0].text
#         print(content)
    except:
        content=" "
    tags=re.findall("#[^\s]+", content)
    date=soup.select("time._1o9PC.Nzb55")[0]['datetime'][:10]
    like=soup.select("div.Nm9Fw > button")[0].text
    place=soup.select("div.M30cS")[0].text
    
    data=[content, date, like, place, tags]
    return data
    
get_conten(driver)
```
